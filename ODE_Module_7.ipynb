## **7. 1 Definisi Deret Fourier**

Pandang suatu fungsi $f$ yang periodik dengan periode $T=2L$ dan kontinu bagian demi bagian (*piecewise continuous*) pada selang $[-L,L]$.

Untuk setiap $t\in \mathbb{R}$, deret Fourier dari $f$ diberikan sebagai 
$$
\dfrac{a_0}{2} + \sum_{n=1}^\infty \left(a_n \cos\left(\dfrac{n\pi t}{L}\right) + b_n\sin\left(\dfrac{n\pi t}{L}\right)\right)
$$

dengan koefisien deret Fourier yang dapat ditentukan dengan 
\begin{align*}
a_0 &= \dfrac{1}{L}\int_{-L}^L f(t)\;dt\\
a_n &= \dfrac{1}{L}\int_{-L}^L f(t)\cos\left(\dfrac{n\pi t}{L}\right)\;dt\\
b_n &= \dfrac{1}{L}\int_{-L}^L f(t)\sin\left(\dfrac{n\pi t}{L}\right)\;dt
\end{align*}

Perhatikan bahwa $a_0$ adalah $a_n$ ketika $n=0$. Suku $a_0$ muncul dari suku $\cos$ dari deret Fourier dikarenakan $\cos\left(\dfrac{n\pi t}{L}\right)=1\neq 0$ saat $n=0$. 

> **Mengapa tidak ada $b_0$?**\
 Karena $\sin\left(\dfrac{n\pi t}{L}\right)=0$ saat $n=0$, sehingga keberadaan $b_0$ tidak ada pengaruhnya pada deret Fourier fungsi.

> **Mengapa $a_0$ dituliskan sebagai $\dfrac{a_0}{2}$? Bukan $a_0$ saja?**\
Dengan menuliskan $\dfrac{a_0}{2}$, kita dapat menggunakan formula $a_n$ untuk menghitung $a_0$ dengan mengganti $n=0$ saja\
$\implies$ Anda hanya perlu menghafal formula $a_n$ dan $b_n$ saja


Hampiran $N$ suku dari deret Fourier $f$ adalah penjumlahan $N$ suku pertama tak nol dari deret Fouriernya.

Semakin banyak suku yang digunakan untuk menghampiri deret Fourier, hampirannya akan semakin baik.

Faktanya, deret Fourier akan tepat konvergen menuju ke fungsi $f$ jika dan hanya jika dilakukan penjumlahan sebanyak tak hingga suku deret Fouriernya, atau
\begin{equation*}
f(x) = \lim_{N\to \infty}\dfrac{a_0}{2} + \sum_{n=1}^N \left(a_n \cos\left(\dfrac{n\pi t}{L}\right) + b_n\sin\left(\dfrac{n\pi t}{L}\right)\right)
\end{equation*}

## **7.2 Deret Fourier Sinus dan Cosinus**

Misalkan diberikan suatu fungsi sebarang $f(t)$ yang terdefinisi pada domain $t\in[0, L]$. Fungsi ini dapat diperluas menjadi fungsi ganjil ataupun fungsi genap.

> Perluasan genap dari $f(t)$ pada $t\in [-L,L]$ yaitu
\begin{equation*}
f_E(t) = \begin{cases}
f(t), &0\leq t\leq L\\
f(-t), &-L<t<0
\end{cases}
\end{equation*}

> Perluasan ganjil dari $f(t)$ pada $t\in [-L,L]$ yaitu
\begin{equation*}
f_O(t) = \begin{cases}
f(t), &0< t< L\\
0, &t=0, L\\
-f(-t), &-L<t<0
\end{cases}
\end{equation*}

Dapat diperiksa dengan cukup mudah bahwa $f_E(t)$ dan $f_O(t)$ yang didefinisikan diatas merupakan fungsi genap dan ganjil, berturut - turut.




> **Deret Fourier Sinus**\
Misalkan $f$ fungsi **ganjil** yang periodik dengan periode $T=2L$ dan kontinu bagian demi bagian pada $[-L,L]$, maka koefisien dari deret Fourier $f$ memenuhi
\begin{align*}
a_n &= 0, \hspace{1cm}n=0, 1, 2, ...\\
b_n &= \dfrac{2}{L}\int_0^L f(x) \sin\left(\dfrac{n\pi x}{L}\right)\;dx, \hspace{1cm}n=1, 2, ...
\end{align*}
dan deret Fourier dari $f$ disebut sebagai Deret Fourier Sinus.

> **Deret Fourier Cosinus**\
Misalkan $f$ fungsi **genap** yang periodik dengan periode $T=2L$ dan kontinu bagian demi bagian pada $[-L,L]$, maka koefisien dari deret Fourier $f$ memenuhi
\begin{align*}
a_n &= \dfrac{2}{L}\int_0^L f(x) \cos\left(\dfrac{n\pi x}{L}\right)\;dx, \hspace{1cm}n=0, 1, 2, ...\\
b_n &= 0, \hspace{1cm}n=1, 2, ...
\end{align*}
dan deret Fourier dari $f$ disebut sebagai Deret Fourier Cosinus.

## **7.3 Implementasi Pada Python**

> ***Contoh 1***\
Tentukanlah deret Fourier dari fungsi $f$ yang didefinisikan sebagai berikut
$$
f(x) = 
\begin{cases}
-x, &-2\leq x<0\\
x, &0\leq x<2
\end{cases}
$$
dengan $f(x+4)=f(x)$. 


Artinya, fungsi $f$ periodik dengan periode $4$. Maka $L=2$. Koefisien deret Fourier dari $f$ yaitu

* Untuk $a_n$, tinjau untuk $n=0$ dan $n>0$ 
    * Untuk $n=0$, diperoleh 

    \begin{align*}
    a_0 &= \dfrac{1}{2}\int_{-2}^2 f(x)\;dx \\
    &= \dfrac{1}{2}\int_{-2}^0 (-x)\;dx + \dfrac{1}{2}\int_{0}^2 x\;dx = 2
    \end{align*}
    * Untuk $n>0$, didapatkan 

    \begin{align*}
    a_n &= \dfrac{1}{2}\int_{-2}^2 f(x)\cos\left(\dfrac{n\pi x}{2}\right)\;dx \\
    &= \dfrac{1}{2}\int_{-2}^0 (-x)\cos\left(\dfrac{n\pi x}{2}\right)\;dx + \dfrac{1}{2}\int_{0}^2 x\cos\left(\dfrac{n\pi x}{2}\right)\;dx\\
    &= \dfrac{4}{(n\pi)^2}(\cos(n\pi)-1)\\
    &= 
    \begin{cases}
    -\dfrac{8}{(n\pi)^2}, &n \text{ ganjil}\\
    0, &n \text{ genap}
    \end{cases}
    \end{align*}
* Untuk $b_n$, diperoleh
\begin{align*}
b_n &= \dfrac{1}{2}\int_{-2}^2 f(x)\sin\left(\dfrac{n\pi x}{2}\right)\;dx \\
&= \dfrac{1}{2}\int_{-2}^0 (-x)\sin\left(\dfrac{n\pi x}{2}\right)\;dx + \dfrac{1}{2}\int_{0}^2 x\sin\left(\dfrac{n\pi x}{2}\right)\;dx\\
&=0
\end{align*}

Jadi, kita peroleh deret Fourier untuk $f$ adalah
\begin{equation*}
f(x) = 1 - \dfrac{8}{\pi^2}\sum_{n\text{ ganjil}}^\infty \dfrac{\cos\left(\dfrac{n\pi x}{2}\right)}{n^2}
\end{equation*}
atau bisa dituliskan juga sebagai
\begin{equation*}
f(x) = 1 - \dfrac{8}{\pi^2}\sum_{n=1}^\infty \dfrac{\cos\left(\dfrac{(2n-1)\pi x}{2}\right)}{(2n-1)^2}
\end{equation*}

*Contoh 1* diatas merupakan salah satu contoh deret Fourier Cosinus.

Perhitungan deret Fourier secara manual menggunakan tangan cukup melelahkan, bukan?

Pada praktikum ini, kita akan menggunakan package `fourier` dari `sympy.series` untuk mempercepat proses di atas.

Untuk fungsi yang didefinisikan bagian demi bagian, tuliskan $f(x)$ sebagai penjumlahan fungsi Heaviside $H(x)=\begin{cases}1, &x\geq0\\0, &x< 0\end{cases}$ atau fungsi unit tangga satuan $u(x-c)$. 

>**Tips Menulis Fungsi Piecewise sebagai Penjumlahan Fungsi Heaviside/Fungsi Tangga Satuan!**\
Gunakan fungsi indikator sebagai berikut:
\begin{align*}
\mathbb{I}_{\{x\geq a\}}&=\begin{cases}1, &x\geq a\\ 0, &\text{$x$ lain}\end{cases}=H(x-a)\\
\mathbb{I}_{\{x> a\}}&=\begin{cases}1, &x> a\\ 0, &\text{$x$ lain}\end{cases}=1-H(a-x)\\
\mathbb{I}_{\{x\leq a\}}&=\begin{cases}1, &x\leq a\\ 0, &\text{$x$ lain}\end{cases}=H(a-x)\\
\mathbb{I}_{\{x< a\}}&=\begin{cases}1, &x< a\\ 0, &\text{$x$ lain}\end{cases}=1-H(x-a)\\
\mathbb{I}_{\{a< x\leq b\}}&=\begin{cases}1, &a< x\leq b\\0, &x \text{ lainnya} \end{cases}=\mathbb{I}_{\{x> a\}}-\mathbb{I}_{\{x> b\}}\\
\mathbb{I}_{\{a\leq x<b\}}&=\begin{cases}1, &a\leq x< b\\0, &x \text{ lainnya} \end{cases}=\mathbb{I}_{\{x\geq a\}}-\mathbb{I}_{\{x\geq b\}}\\
\mathbb{I}_{\{a\leq x\leq b\}}&=\begin{cases}1, &a\leq x\leq b\\0, &x \text{ lainnya} \end{cases}=\mathbb{I}_{\{x\geq  a\}}-\mathbb{I}_{\{x> b\}}\\
\mathbb{I}_{\{a< x< b\}}&=\begin{cases}1, &a< x< b\\0, &x \text{ lainnya} \end{cases}=\mathbb{I}_{\{x>  a\}}-\mathbb{I}_{\{x\geq b\}}
\end{align*}
untuk setiap $a,b\in \mathbb{R}, 0\leq a<b$.

Tanda `=` hanya perlu diperhatikan secara serius apabila Anda hendak menggunakan fungsi `f` untuk keperluan plotting dsb. Apabila `f` hanya didefinisikan untuk menghitung deret Fourier, Anda dapat langsung menganggap semua interval `a<x<b`, `a<=x<=b`, `a<x<=b` sebagai `a<=x<b`. (Karena `a<=x<b` lebih sederhana dan mudah dihafal)


Kembali lagi ke Contoh 1. Diperoleh
\begin{equation*}
f(x) =  -xH(x+2) - xH(x-2)+2xH(x)
\end{equation*}

atau dapat juga kita langsung gunakan $f(x)=|x|$.









```python
from sympy import *
from sympy.abc import x
# Dengan f(x) = -xH(x+2) - xH(x-2) +2xH(x)
fseries_f = fourier_series(-x*Heaviside(x+2)-x*Heaviside(x-2)+2*x*Heaviside(x), (x, -2, 2))
fseries_f
```

```python
# Dengan f(x)=|x|
fseries_f = fourier_series(abs(x), (x, -2, 2))
fseries_f
# (Keduanya memberikan hasil deret Fourier yang sama)
```

`fseries_f` yang dihasilkan pada blok kode diatas merupakan objek `FourierSeries` yang memiliki beberapa atribut penting sebagai berikut,


1.   `a0`, `an`, `bn` : koefisien $a_0, a_n$ dan $b_n$ dari deret Fourier
2.   `period`  : tupel berisikan domain dari deret Fourier
3.   `truncate(n)` : $n$ - suku tak nol pertama dari deret Fourier


```python
print(f"Koefisien a_0 : {fseries_f.a0}")
```

```python
print("Koefisien a_n : \n")
fseries_f.an
```

```python
print("Koefisien b_n : \n")
fseries_f.bn
```

```python
print("Domain deret Fourier f : \n")
fseries_f.period
```

```python
print("10 Suku tak nol pertama deret Fourier f : ")
fseries_f.truncate(10)
```

Selanjutnya, akan ditampilkan plotting dari deret Fourier tersebut.


```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 120 # Mengatur dpi figure agar gambar lebih tajam.
f2 = lambdify(x, fseries_f.truncate(2))
f3 = lambdify(x, fseries_f.truncate(3))
f5 = lambdify(x, fseries_f.truncate(5))
f = lambda x : abs(x)
xs = np.linspace(-2, 2, 1000)
plt.plot(xs, f(xs), 'k-')
plt.plot(xs, f2(xs), 'r-')
plt.plot(xs, f3(xs), 'g-')
plt.plot(xs, f5(xs), 'b-')
plt.legend(['f(x)', 'n=2', 'n=3', 'n=5'])
plt.show()
```

> ***Contoh 2***\
Tentukanlah deret Fourier dari fungsi $g(x)=u(x)$ pada domain $[-5, 5]$, yang periodik dengan $g(x+10) = g(x)$. Tentukanlah hampiran deret Fourier tersebut untuk 2, 5, 7 dan 10 suku pertama. Plot perbandingan antara $g(x)$ dengan hampiran deret Fouriernya pada 4 *figure* yang berbeda.

```python
# Menentukan deret Fourier g.
from sympy import *
from sympy.abc import x
fseries_g = fourier_series(Heaviside(x), (x, -5, 5))
fseries_g
```

```python
print(f"Koefisien a_0 : {fseries_g.a0}")
```

```python
print("Koefisien a_n : \n")
fseries_g.an
```

```python
print("Koefisien b_n : \n")
fseries_g.bn
```

```python
print("Period deret Fourier g : \n")
fseries_g.period
```

```python
print("10 Suku tak nol pertama deret Fourier g : ")
fseries_g.truncate(10)
```

```python
g = lambda x : np.heaviside(x, 0.5)
# Plotting
xs = np.linspace(-5, 5, 500)
fig, ax = plt.subplots(2, 2, figsize=(6, 5))
ord = np.array([[2, 5], [7, 10]]) # Banyak suku hampiran
for i in range(2):
    for j in range(2):
        gN = fseries_g.truncate(ord[i, j]) # Tentukan hampiran deret Fourier untuk 2,5,7,10 suku tak nol pertama.
        gN_funcs = lambdify(x, gN)         # Convert hampiran deret Fourier menjadi fungsi
        ax[i, j].plot(xs, g(xs))           # Plot fungsi g
        ax[i, j].plot(xs, gN_funcs(xs))    # Plot hampiran N suku deret Fourier g 
        ax[i, j].set(title="Hampiran " + str(ord[i, j])+ " suku")
        ax[i, j].grid('on')
plt.tight_layout() # Merapikan layout akhir.
```

**Life Hack** : Untuk mendefinisikan fungsi periodik, Anda dapat manfaatkan sifat berikut.
> Misalkan fungsi $f(x)$ periodik dengan periode $T$ di interval $(a, b)$ maka $f(x) = f((x-a)\mod T + a)$ untuk setiap $x\in\mathbb{R}$.

Contoh, misalkan $f(x)=x^2+4$ di $(-2,2)$ yang periodik dengan periode $4$. \
Maka fungsi $f$ dapat dituliskan sebagai 
$$
f(x) = f([(x+2)\mod 4] -2)=([(x+2)\mod 4] -2)^2 +4
$$
untuk setiap $x\in \mathbb{R}$.

```python
xs = np.linspace(-12, 12, 500)
f = lambda x : x**2 + 4
fin = lambda x : (x + 2)%4 - 2
plt.plot(xs, f(fin(xs)))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid('on')
```
